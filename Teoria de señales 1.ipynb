```
import numpy as np 
from matplotlib import pyplot as plt
import math
```

```
t = np.linspace (0,2, 1000)
y1 = np.sin (2*np.pi*3*t)
y2 = np.cos (2*np.pi*3*t)
#t = np.arange (0 , 10 , 0.1) otra forma de crear un arreglo
```

```
y1.shape
```

```
plt.figure(figsize=(9,4), dpi=90) #Ajustar tama√±o del grafico

plt.plot(t, y1, label='sin')
plt.plot(t, y2, label='cos')

plt.legend() #activar lable 
plt.grid(True) #negrilla

plt.title('Sin and Cos') #titulo
plt.xlabel('Time[$t$]')  #titulo x eje
plt.ylabel('Amplitude')  #titulo y eje
```
