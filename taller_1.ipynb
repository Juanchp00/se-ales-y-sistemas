#2.Conceptos básicos de señales 
Evaluar la expresión:
\begin{equation}
    \int_{-\infty}^{\infty} t^2 \cdot e^{-sin(t)} \cdot cos(2t) \cdot \delta(2t-2π)dt
\end{equation}


```
%matplotlib inline
import matplotlib.pyplot as plt
import sympy as sym
from sympy import init_printing
import numpy as np

t=sym.symbols('t', real=True)
pi=np.pi

ec=(t**2)*sym.exp(-sym.sin(t))*sym.cos(2*t)*sym.DiracDelta(2*t-2*pi)

rta=sym.integrate(ec, (t, -sym.oo, sym.oo))
print (rta)
```

Sea $x(t)=u(t-t_o)-u(t-nt_o)-k \delta(t-mt_o)$. Determine el valor de k para el cual:
\begin{equation}
    \int_{-\infty}^{\infty} x(t)dt = 0
\end{equation}

```
%matplotlib inline
import matplotlib.pyplot as plt
import sympy as sym
from sympy import init_printing
import numpy as np
m,n,to,t=sym.symbols('m n to t', real=True)
k=to*(n-1)
u = sym.Heaviside(t)
unto = sym.Heaviside(t-n*to)
st = u -unto
delta= sym.DiracDelta(t-m*to)
ec=st-k*delta
ec1=st-k
y=sym.integrate(ec1, (to, n*to))
print (y)

```

* Hallar la salida y[n] de un SLIT ante la entrada $x[n]=${$−1, 1, 3^†, 0, 5, 6, −1$}, con respuesta al impulso $h[n]=${$1, −2, 0^†, 1, −1$}, donde $n\in${$0, ±1, ±2, . . . , ±N$} y $n=0$ para $x[n]^†$. * Sea la señal Gaussiana $x(t) = e^{−at^{2}}$
con $a \in \mathbb {R}+$, el sistema A con relacion entrada-salida $y_A(t) = x^2(t)$, y el sistema lineal e invariante con el tiempo B con respuesta al impulso $h_B(t) = Be^{−bt^{2}}$:

  a) Encuentre la salida del sistema en serie $x(t) → h_B(t) → y_A(t) → y(t)$ 
  
  b) Encuentre la salida del
sistema en serie $x(t) → y_A(t) → h_B(t) → y(t)$.



```
#simulación señal a)
import numpy as np
n=5
x1=np.sin(50*np.pi*n)
xo1=np.sin(50*np.pi*(n-1))
x2=np.cos(12*np.pi*n)
xo2=np.cos(12*np.pi*(n-1))
#linealidad
m =0
l=0
a=3
b=2
for h in range (0,(n-1)):
  xp1=np.sin(50*np.pi*h)
  xpo1=np.sin(50*np.pi*(h-1))
  xp2=np.cos(12*np.pi*h)
  xpo2=np.cos(12*np.pi*(h-1))
  y1=((a*xp1+b*xp2)/3)+2*(a*xpo1+b*xpo2)-m
  y2=a*((xp1/3)+(2*xpo1))+b*((xp2/3)+(2*xpo2))-l
  m=y1
  l=y2

yp=((a*x1+b*x2)/3)+2*(a*xo1+b*xo2)-y1
ys=a*((x1/3)+(2*xo1))+b*((x2/3)+(2*xo2))-y2
print('el sistema es lineal pues el primer caso de y1=',yp,'es de igual valor al segundo caso de y2=',ys)

#invariabilidad en el tiempo
no=3
lim=n-no
xp1=np.sin(50*np.pi*lim)
xpo1=np.sin(50*np.pi*(lim-1))
m=0
for h in range (0,(lim-1)):
  xz1=np.sin(50*np.pi*h)
  xzo1=np.sin(50*np.pi*(h-1))
  y1=((xz1)/3)+2*(xzo1)-m
  m=y1

ya=((xp1/3)+2*xpo1)-y1
yb=(np.sin(50*np.pi*(lim))/3)+2*np.sin(50*np.pi*(lim-1))-y1
    
print('el sistema es invariable pues el primer caso de y[n-no]=',ya,'es de igual valor al segundo caso de y[n;no]=',yb)


```

```
#simulación señal b)
import numpy as np
import sympy as sym
from sympy import init_printing
#podemos ver que este sistema es invertible y su inverso es x[n]-x[n-1], si tomamos a x[n]como delta[n], observamos que delta[n] seria la respuesta al impulso h[n],por lo que para encontrar la salida del sistema deberiamos hacer una convolucion de la respuesta al impulso y una función x[n]
A=9
B=6
n=8
h=sym.DiracDelta(n)
x=A*np.sin(12*np.pi*n)

#linealidad
a=3
b=3
x1=A*np.sin(12*np.pi*n)
x2=B*np.cos(22*np.pi*n)

y1=np.convolve((a*x1+b*x2),h)
y2=a*np.convolve(x1,h)+b*np.convolve(x2,h)

print('el sistema es lineal pues el primer caso de y, y1=',y1,'es igual a segundo caso de y, y2=',y2)

#invariabilidad en el tiempo
no=3
lim=n-no
xa=A*np.sin(12*np.pi*lim)

ya=np.convolve(xa,h)
yb=np.convolve((A*np.sin(12*np.pi*lim)),h)

print('el sistema es invariable pues el primer caso de y[n-no]=',ya,'es de igual valor al segundo caso de y[n;no]=',yb)
```
